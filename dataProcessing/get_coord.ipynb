```python
import pymongo
myclient = pymongo.MongoClient("mongodb://finder:finder@irinka.cn:26926/homefinder")
o=open("./missingCount","w")
mydb = myclient["homefinder"]
lines=f.readlines()
for i in lines:
    blockName=i[0:-1]
    counts=mydb["sales_records"].count_documents({"block":blockName})
    print(blockName+": "+str(counts))
    o.write(blockName+": "+str(counts)+"\n")

o.close()
```

```python
lst=open("./missingCount","r")

lines=lst.readlines()
n_lst=[]
for line in lines:
    kv=line.split(': ')
    num=kv[1].lstrip()
    num=num.rstrip()
    num=int(num)

    n_lst.append({"block":kv[0],"num":num})

def key(el):
    return el["num"]

n_lst.sort(key=key,reverse=True)
```

```python
o=open("./blockWithCount","w")

for kv in n_lst:
    o.write(kv['block']+": "+str(kv['num'])+'\n')

o.close()
```
