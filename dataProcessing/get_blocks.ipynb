```python
import pymongo
myclient = pymongo.MongoClient(
    "mongodb://finder:finder@irinka.cn:26926/homefinder")
mydb = myclient["homefinder"]
ans=mydb["sales_records"].aggregate([
    {'$group': {
        '_id': "$block",
        'block': {'$first': "$block"},
        'year': {'$max': "$year"},
        'month': {'$max': "$year", '$max': "$month"},
    },

    },
    {'$project': {
        '_id': 0
    }
    }
])
ans
```

```python
out=[]
for i in ans:
    print(i)
    out.append(i)
```

```python
import json
region_json = open("./block_data.json", 'r')
js_str = region_json.read()
# print(js_str)

js = json.loads(js_str)
# print(js)

region = {}

for i in js:
    region[i['block']] = i

print(region)
of = open("./blocks", 'w')
for i in out:
    if i['block'] in region:
        r = region[i['block']]['region']
        sr = region[i['block']]['sub_region']
        a = '北京 '+r+' '+sr+' '+i['block']+'\n'
    else:
        a = '北京 '+i['block']+'\n'

    # print(a)
    of.write(a)
of.close()
```
