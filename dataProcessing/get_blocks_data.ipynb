```python
# from coordTransform_utils import *
import pandas as pd

file1 = './map-location.xlsx'
file2 = './GCJ-map-location.xlsx'
file3 = './GCJ_occrected_map-location.xlsx'

try:
    data3 = pd.read_excel(file3, engine='openpyxl')
    data2 = pd.read_excel(file2, engine='openpyxl')
    data1 = pd.read_excel(file1, engine='openpyxl')
except Exception as e:
    raise e
```

```python

from coordTransform_utils import *
```

```python
valid_data1 = data1[data1['错误'].isnull() == True].copy()
valid_data1['lng,lat'] = valid_data1.apply(
    lambda row: str(bd09_to_wgs84(row['经度'], row['纬度'])), axis=1)
```

```python
def parse_lat_lng(row):
    lag = row['纬度,经度'].split(',')
    return str(gcj02_to_wgs84(float(lag[0]),float(lag[1])))
```

```python
print(data2[data2['纬度,经度'].isna() == True])
# valid_data2
valid_data2 = data2[data2['纬度,经度'].isna() == False].copy()

valid_data2['lng,lat'] = valid_data2.apply(
    lambda row: parse_lat_lng(row), axis=1)

valid_data2
```

```python
valid_data3 = data3.copy()

valid_data3['lng,lat'] = valid_data3.apply(
    lambda row: parse_lat_lng(row), axis=1)

valid_data3['地址']= valid_data3.apply(
    lambda row: row['地址'][3:], axis=1)

valid_data3
```

```python
import pymongo
myclient = pymongo.MongoClient("mongodb://finder:finder@irinka.cn:26926/homefinder")
mydb = myclient["homefinder"]
def get_region(block):
    region=mydb["sales_records"].find_one({"block":block})
    return (region['region'],region['sub_region'])
```

```python
# print(valid_data1)
# print(valid_data2)
# print(valid_data3)
region_dic={}
for i in valid_data1['地址']:
    ans=get_region(i)
    region_dic[i]={'region':ans[0],'sub_region':ans[1]}
    print(region_dic[i])
```

```python
for i in valid_data2['地址']:
    ans=get_region(i)
    region_dic[i]={'region':ans[0],'sub_region':ans[1]}
    print(region_dic[i])

for i in valid_data3['地址']:
    ans=get_region(i)
    region_dic[i]={'region':ans[0],'sub_region':ans[1]}
    print(region_dic[i])
```

```python
import re
ans = []


def parseLA(lng_lat):
    matchObj = re.match('\[(.*), (.*)\]', lng_lat, re.M | re.I)
    return float(matchObj.group(1)), float(matchObj.group(2))

def addItem(row):
    lng, lat = parseLA(row['lng,lat'])
    addr=row['地址']
    ans.append({'block':addr , 'region': region_dic[addr]['region'], 'sub_region': region_dic[addr]['sub_region'], 'lat': lat, 'lng': lng})

valid_data1.apply(lambda row:addItem(row),axis=1)
valid_data2.apply(lambda row:addItem(row),axis=1)
valid_data3.apply(lambda row:addItem(row),axis=1)

print(len(ans))
```
